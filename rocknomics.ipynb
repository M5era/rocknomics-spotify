### Why is predicting the popularity of a song important for Spotify?

Spotify has a platform-based business model that works with three key stakeholders: 
1. Artists that publish their songs on the platform
2. Users that listen to music and podcasts on the platform either by paying for it or by listening to advertisements
3. Advertisers that pay Spotify to put advertisements on the platform

Therefore, it is relevant for Spotify to go informed into the negotiations with artists and labels. Additionally, advertisers are a major revenue stream for Spotify. A model that can predict popularity can support Spotify in the negotiations with advertisers and better allocate relevant advertisements after popular songs. 

Most importantly, Spotify constantly creates new relevant playlists to keep its users engaged. By predicting the popularity of a song, Spotify can place new popular songs in algorithmic playlists like Release Radar and Discover Weekly, to update its users with the best new songs. This recommendation engine is vital for Spotify’s success, as it helps their users discover new tracks and keeps them engaged with the platform.

With our model, we want to provide an accurate prediction of popularity based on easily obtainable data of the song, which can be acquired without any further research. This model then can help Spotify to go well informed into price negotiations with artists and labels, optimize advertising algorithms, and update and optimize playlists with good songs that keep their users engaged.

Moreover, song popularity is hard to detect by rules and the history of individual songs and artists might not be that long. Therefore, machine learning will be able to help us answer our question and to predict song popularity.



2.) Secondly, we need to formulate the problem. Ultimately, our goal is to predict a song's popularity. We will choose these most relevant features for our model through feature selection and analyzing their impact on the popularity score.



3.) Thirdly, the relevant data has to be collected. We will base our analysis on five datasets which contain information about 175.000 songs which have been published between 1921-2020. The data was collected from the Spotify Web API, and contains data grouped by artist, year, or genre in the data section, as well as audio features and track features. The size of the data pool for training our model therefore is large enough to perform a significant prediction (> 50).



4.) In the fourth step, we need to ask if our problem has regular patterns. Probably, the songs from past years will show patterns, for instance, with regard to famous artists and popular genres. 5.) In step five, we need to look for meaningful representations and features of the data. ### Variable description:

- **duration:** The duration of the track in milliseconds.
- **key:** The estimated overall key of the track. Integers map to pitches using standard Pitch Class notation . E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on.
- **audio_mode:** Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
- **time_signature:** An estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).
- **acousticness:** A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
- **danceability:** Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
- **energy:** Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.
- **instrumentalness:** Predicts whether a track contains no vocals. “Ooh” and “aah” sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly “vocal”. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.
- **liveness:** Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
- **loudness:** The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typical range between -60 and 0 db.
- **speechiness:** Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.
- **audio_valence:** A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
- **tempo:** The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

#### Preprocessing Pipeline

```python
# preprocessing: OneHotEncoding categorical, binning binary, and standardizing numeric variables

cat_attribs = ['key']
num_attribs = ['acousticness', 'danceability', 'energy','instrumentalness', 'liveness', 'speechiness', 
               'valence','duration_min', 'tempo', 'loudness', 'year', 'artist_popularity'] 
bin_attribs = ['explicit', 'mode']

def do_nothing(df_in):
    return df_in

cat_pipeline = Pipeline([
        ('cat_encoder', OneHotEncoder(sparse=False))
    ])

bin_pipeline = Pipeline([
    ('binary', FunctionTransformer(do_nothing))
])

num_pipeline = Pipeline([
        ('std_scaler', StandardScaler())
    ])

preprocessing_pipeline = ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown='ignore'), cat_attribs),
        ("bin", bin_pipeline, bin_attribs),
        ("num", num_pipeline, num_attribs),
    ])
```

```python
preprocessing_pipeline
```

```python
# splitting data into train and test sets

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)

folds = 5
```

### 3.1. Linear Regression

```python
# linear regression with GridSearch cross-validation 


lin_model_pipeline = Pipeline([
    ('preprocessor', preprocessing_pipeline),
    ('lin', LinearRegression())
])

lin_grid_pipeline = Pipeline([
    ('preprocessor', preprocessing_pipeline),
    ('grid', GridSearchCV(LinearRegression(), cv=folds, n_jobs=-1, verbose=5,
                    param_grid ={
                    'fit_intercept':[True, False],
                    'normalize': [True, False]}
                    ))
])

lin_grid_pipeline.fit(X_train,y_train)
```

```python
X_train_preprocessed = preprocessing_pipeline.fit_transform(X_train)
X_test_preprocessed = preprocessing_pipeline.fit_transform(X_test)
```

```python
print('Test R^2 Score : %.3f'%lin_grid_pipeline['grid'].best_estimator_.score(X_test_preprocessed,y_test))
print('Best R^2 Score Through Grid Search (cv-score) : %.3f'%lin_grid_pipeline['grid'].best_score_)
print('Best Parameters : ',lin_grid_pipeline['grid'].best_params_)
```

```python
# Lasso regression with GridSearch cross-validation 


lasso_grid_pipeline = Pipeline([
    ('preprocessor', preprocessing_pipeline),
    ('grid', GridSearchCV(Lasso(random_state=42), cv=folds, n_jobs=-1, verbose=5,
                    param_grid ={
                    'alpha':[0.005, 0.01, 0.05, 0.1, 0.3, 0.6, 0.8, 1]},
                    scoring='r2',
                    ))
])

lasso_grid_pipeline.fit(X_train, y_train)

```

```python
print('Test R^2 Score : %.3f'%lasso_grid_pipeline['grid'].best_estimator_.score(X_test_preprocessed,y_test))
print('Best R^2 Score Through Grid Search (cv-score) : %.3f'%lasso_grid_pipeline['grid'].best_score_)
print('Best Parameters : ',lasso_grid_pipeline['grid'].best_params_)
```

#### Feature importance of model

```python
# display importance by feature for the two best models 

def showFeatureImportance(grid_pipeline):

    importance = grid_pipeline['grid'].best_estimator_.feature_importances_
    list_importance = []
    for i,v in enumerate(importance):
        list_importance.append([int(i),round(v,5)])

    df_imp = pd.DataFrame(data=list_importance, columns =['feature','weight'])
    df_imp[df_imp['weight']>0.005]

    cat_features = preprocessing_pipeline.named_transformers_['cat'].get_feature_names()
    feature_names = list(cat_features) + bin_attribs + num_attribs
    
    feature_indices = df_imp.sort_values(by='weight', ascending= False).index.to_numpy()
    
    #Add feature names to the dataFrame
    df_imp['feature_name'] = ""
    for ix in feature_indices:
        df_imp.iloc[ix,2] = feature_names[ix]
    display(df_imp.sort_values(by='weight', ascending= False).head(10))
```

The second most influential feature is the year with roughly 4.84 %, followed by the duration in min with 1.1%. 

Liveness and loudness have the 4th and 5th highest influence with 1 %, and 0.99 % respectively. Looking back at the correlations we can say that midrange loudness, and songs recorded with an audible audience seem to do worse and predict a song as unpopular by a small percentage.

    

## 5. Conclusion

For Spotify, this implies that the popularity of the artist, as well as the year of a song's release, and its duration, should be the primary research factors when going into negotiations with an artist and labels. Likewise, these three factors can support Spotify to better allocate relevant advertisements prior to a song's release, and increase profits off of higher ad revenues. Playlists and song recommendations can also be improved by including songs that are predicted to be more popular so that the statistical probability that the user will enjoy the songs increases.

From the second most important feature 'year' Spotify can not draw strong conclusions. Songs decrease popularity over time, so more current releases tend to be more hyped (popular) compared to songs released decades ago. However, that does not mean, that is song was not popular around its release date. This can also be seen as a drawback of our model as the data is only a short snapshot of time. It would be interesting to further analyze the importance of these features over time. 

An important note in our findings is that other features could have high explanatory power, such as record labels, which can lead to omitted variable bias. While artist popularity is usually a good predictor for a song's popularity we believe that more data can be included in the model for more accurate and revelatory outcomes. Furthermore, we are only predicting popularity as rated by Spotify, and only using data of their platform. Our predictions therefore might not be transferable to other populations, since the average Spotify user is statistically likely to be demographically and culturally different from other music enjoyers.

Since the most important feature of the model is an artist's current popularity, it might be very difficult to predict newcomer artists, which is especially relevant for the artist negotiations that Spotify has. We recommend Spotify to include factors like social media followings and views in their analysis, as these might be good parameters to predict the popularity of an artist, and thereby also the predicted popularity of their new songs.
